Resizing the image by using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/Alaskan-Malamute-standing-in-the-grass1-400x267.jpg')
img_resized = img.resize((200, 200))
```

```python
img_resized.save('dog_image_resized.jpg')
```

```python
img_res = mpimg.imread('/content/dog_image_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
print(img_res.shape)
```

Converting RGB images to Grayscale image by using OpenCV

```python
import cv2
```

```python
img = cv2.imread('/content/Alaskan-Malamute-standing-in-the-grass1-400x267.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(grayscale_img)
```

```python
cv2.imwrite('dog_grayscaleImg.jpg', grayscale_img)
```
